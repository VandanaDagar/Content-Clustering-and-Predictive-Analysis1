```python
#1. How many number of movies and tv-shows are there?

import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the bar plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='type', palette=['skyblue', 'lightgreen'])

# Adding labels and title
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Number of Movies vs. TV Shows')

# Show plot
plt.show()

```

```python
#How many movies and tv-shows were released per year?

# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'])

# Extract year and month from 'date_added'
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Group by year and count the number of entries per year
yearly_trend = df.groupby('year_added').size()

# Group by year and month for more granular analysis
monthly_trend = df.groupby(['year_added', 'month_added']).size()

#Plotting graph
plt.figure(figsize=(10, 6))
yearly_trend.plot(kind='line', marker='o', linestyle='-')
plt.title('Number of Movies/TV Shows Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Additions')
plt.grid(True)
plt.show()
```

```python
#Number of Movies/TV Shows Added Per Month Over the Years.

monthly_trend = monthly_trend.unstack(level=0)  # Reshape for plotting

plt.figure(figsize=(14, 7))
monthly_trend.plot(kind='line', marker='o')
plt.title('Number of Movies/TV Shows Added Per Month Over the Years')
plt.xlabel('Month')
plt.ylabel('Number of Additions')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Distribution of ratings among the movies and tv-shows.

# Filter for movies and TV shows separately (optional)
movies_df = df[df['type'] == 'Movie']
tv_shows_df = df[df['type'] == 'TV Show']

# Count ratings for movies
movie_ratings_counts = movies_df['rating'].value_counts()

# Count ratings for TV shows
tv_show_ratings_counts = tv_shows_df['rating'].value_counts()

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot count plot for movies
sns.countplot(x='rating', data=movies_df, order=movies_df['rating'].value_counts().index)
plt.title('Distribution of Ratings Among Movies')
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()

# Plot count plot for TV shows
plt.figure(figsize=(12, 6))
sns.countplot(x='rating', data=tv_shows_df, order=tv_shows_df['rating'].value_counts().index)
plt.title('Distribution of Ratings Among TV Shows')
plt.xlabel('Rating')
plt.ylabel('Number of TV Shows')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming 'listed_in' is the column containing the categories

# Split the categories in each row and expand into a new DataFrame
categories_expanded = df['listed_in'].str.split(',', expand=True)

# Stack the DataFrame to convert it from a wide format to a long format
categories_stacked = categories_expanded.stack()

# Strip any leading or trailing whitespace characters from the categories
categories_stacked = categories_stacked.str.strip()

# Get unique categories by converting the Series to a set
unique_categories = set(categories_stacked)

# Convert the set back to a list (optional, if you prefer list format)
unique_categories_list = list(unique_categories)

# Print the unique categories
print(unique_categories_list)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
# df = pd.DataFrame({'listed_in': ['Drama, Action', 'Comedy, Drama', 'Action, Sci-Fi', 'Drama, Action', 'Comedy']})

# Step 1: Split the 'listed_in' column into individual categories
categories_expanded = df['listed_in'].str.split(',', expand=True)

# Step 2: Stack the categories to a single column and strip leading/trailing spaces
categories_stacked = categories_expanded.stack().str.strip()

# Step 3: Count the occurrences of each category
category_counts = categories_stacked.value_counts()

# Display the category counts (Optional, for verification)
print(category_counts)
```

```python
#Number of movies per category.
# Plot the data
plt.figure(figsize=(10, 8))
category_counts.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Category')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```
